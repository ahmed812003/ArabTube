```python
import cv2
import os
import shutil
import numpy as np
from tensorflow.keras.models import load_model
```

TEST MODEL

```python
model = load_model("/content/drive/MyDrive/best_model.keras")
```

```python
# Function to preprocess frames
def preprocess_frame(frame):
    # Resize the frame to match the input size of the model
    frame = cv2.resize(frame, (299, 299))
    # Rescale pixel values to the range [0, 1]
    frame = frame / 255.0
    return frame
```

```python
def classify_frame(frame):
    # Preprocess the frame
    processed_frame = preprocess_frame(frame)
    # Add batch dimension as the model expects batches of images
    processed_frame = np.expand_dims(processed_frame, axis=0)
    # Predict using the model
    prediction = model.predict(processed_frame)
    # Return the prediction
    return prediction
```

```python


def extract_frames(video_path, threshold=0.8):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    sexual_frames , non_sexual_frames = 0 , 0
    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            break

        prediction = classify_frame(frame)
        # Print the prediction for debugging purposes
        #print(f"Frame {i}: Prediction = {prediction}")

        # Check if the prediction indicates pornography
        if prediction[0][0] > threshold:
            # Save the frame to the porn directory
            #frame_filename = os.path.join(output_dir_porn, f"frame_{i}.jpg")
            sexual_frames += 1
        else:
            # Save the frame to the non-porn directory
            #frame_filename = os.path.join(output_dir_non_porn, f"frame_{i}.jpg")
            non_sexual_frames += 1
        #cv2.imwrite(frame_filename, frame)

    cap.release()
    return sexual_frames , non_sexual_frames
```

```python
video_path = r"/content/test.mp4"
sexual_frames, non_sexual_frames = extract_frames(video_path)
total_frames = sexual_frames + non_sexual_frames
print(f"Number of total_frames = {total_frames}")
print(f"Number of sexual_frames = {sexual_frames}")
print(f"Number of non_sexual_frames = {non_sexual_frames}")
print(f"Percision of a video as sexual video = {(sexual_frames / total_frames) * 100}")
print(f"Percision of a video as non sexual video = {(non_sexual_frames / total_frames) * 100}")
```
